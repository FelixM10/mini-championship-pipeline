# Swansea City 2024/25 – One Page Summary (BigQuery Source)

This notebook builds a one-page style summary for **Swansea City** using
BigQuery tables:

- `player_stats_semantic_2024_25`
- `league_table_enhanced_2024_25`

It shows:

1. Team profile vs league (points, xG, goals, possession, progression, net spend)
2. Swansea player attacking contribution (xG+xAG vs G+A per 90)
3. Swansea ball progression map (progressive passes vs carries per 90)
4. Minutes & role distribution (who actually plays)


```python
import sys
from pathlib import Path
import warnings


# ------------------------------------------------
# Robustly locate the project root so we can import src.config
# Works both:
#   - when running interactively from notebooks/
#   - when run via nbconvert from the project root
# ------------------------------------------------
def find_project_root() -> Path:
    cwd = Path.cwd().resolve()
    candidates = [
        cwd,
        cwd.parent,
        cwd.parent.parent,
    ]

    for candidate in candidates:
        if (candidate / "src" / "config.py").exists():
            return candidate

    raise RuntimeError(
        f"Could not locate project root from {cwd}. "
        "Tried current directory and two parents."
    )


ROOT = find_project_root()
if str(ROOT) not in sys.path:
    sys.path.append(str(ROOT))

from src.config import GCP_PROJECT_ID, BQ_DATASET

from google.cloud import bigquery
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rcParams["figure.figsize"] = (8, 5)
plt.rcParams["axes.grid"] = True
plt.rcParams["font.size"] = 10

client = bigquery.Client(project=GCP_PROJECT_ID)

TABLE_PLAYERS = f"{GCP_PROJECT_ID}.{BQ_DATASET}.player_stats_semantic_2024_25"
TABLE_LEAGUE  = f"{GCP_PROJECT_ID}.{BQ_DATASET}.league_table_enhanced_2024_25"
```

```python
players_query = f"SELECT * FROM `{TABLE_PLAYERS}`"
league_query  = f"SELECT * FROM `{TABLE_LEAGUE}`"

players = client.query(players_query).to_dataframe()
league  = client.query(league_query).to_dataframe()

#print("Players shape:", players.shape)
#print("League shape :", league.shape)

#players.head(), league.head()
```

```python
import pandas as pd
import numpy as np

# Robust Swansea detection in league table
clubs = league["club"].astype(str).str.strip()
matches = [c for c in clubs.unique() if "swansea" in c.lower()]

if not matches:
    raise ValueError("Could not find any club containing 'Swansea' in league table.")

SWANSEA_NAME = matches[0]
#print("Detected Swansea club name:", SWANSEA_NAME)

# Ensure a 'matches played' column
if "played" in league.columns:
    league["matches"] = league["played"]
elif "squad_mp" in league.columns:
    league["matches"] = league["squad_mp"]
else:
    raise ValueError("Could not find a matches/played column in league_table_enhanced.")

def has(col: str) -> bool:
    return col in league.columns

team_metrics: list[tuple[str, str]] = []

# Points per game
if has("pts") and has("matches"):
    league["pts_per_game"] = league["pts"] / league["matches"]
    team_metrics.append(("pts_per_game", "Points per game"))

# xG for per game
if has("squad_xg") and has("matches"):
    league["squad_xg_per_game"] = league["squad_xg"] / league["matches"]
    team_metrics.append(("squad_xg_per_game", "xG for per game"))

# xAG for per game
if has("squad_xag") and has("matches"):
    league["squad_xag_per_game"] = league["squad_xag"] / league["matches"]
    team_metrics.append(("squad_xag_per_game", "xAG for per game"))

# Goals for per game
if has("goals_for") and has("matches"):
    league["gf_per_game"] = league["goals_for"] / league["matches"]
    team_metrics.append(("gf_per_game", "Goals for per game"))

# Goals against per game (lower = better score)
if has("goals_against") and has("matches"):
    league["ga_per_game"] = league["goals_against"] / league["matches"]
    league["ga_prevented_score"] = -league["ga_per_game"]
    team_metrics.append(("ga_prevented_score", "Goals against per game"))

# Possession
if has("squad_poss"):
    team_metrics.append(("squad_poss", "Possession %"))

# Progressive passes p90
if has("squad_prgp") and has("squad_90s"):
    league["squad_prgp_p90"] = league["squad_prgp"] / league["squad_90s"]
    team_metrics.append(("squad_prgp_p90", "Progressive passes p90"))

# Progressive carries p90
if has("squad_prgc") and has("squad_90s"):
    league["squad_prgc_p90"] = league["squad_prgc"] / league["squad_90s"]
    team_metrics.append(("squad_prgc_p90", "Progressive carries p90"))

# Net spend
if has("net_spend_eur"):
    team_metrics.append(("net_spend_eur", "Net spend (€)"))

#print("Team metrics to compare:", team_metrics)
```

```python
# Re-select Swansea row *after* all derived metrics are added
swansea_row = league[league["club"] == SWANSEA_NAME].iloc[0]

rows = []
for col, label in team_metrics:
    if col not in league.columns:
        continue
    sw_val = swansea_row[col]
    league_median = league[col].median()
    ratio = sw_val / league_median if pd.notna(league_median) and league_median != 0 else np.nan
    rows.append(
        {
            "metric_col": col,
            "metric": label,
            "swansea_value": sw_val,
            "league_median": league_median,
            "ratio_vs_median": ratio,
        }
    )

team_df = pd.DataFrame(rows)
team_df
```

```python
plt.figure(figsize=(8, 5))

y_pos = np.arange(len(team_df))
ratios = team_df["ratio_vs_median"].values

plt.axvline(1.0, color="gray", linestyle="--", linewidth=1)

bars = plt.barh(
    y_pos,
    ratios,
    color=np.where(ratios >= 1.0, "#1b9e77", "#d95f02"),
    alpha=0.8,
)

plt.yticks(y_pos, team_df["metric"])
plt.xlabel("Swansea value / League median")
plt.title(f"{SWANSEA_NAME} - Team Profile vs League Median")

for i, r in enumerate(ratios):
    if np.isnan(r):
        continue

    metric = team_df["metric"].iloc[i]
    swan_val = team_df["swansea_value"].iloc[i]

    # Invert Swansea value only for GA metric
    if metric == "Goals against per game":
        swan_val = -swan_val

    if metric == "Net spend (€)":
        label = f"{int(swan_val):,}"
    else:
        label = f"{swan_val:.2f}"

    if r >= 0:
        x_text = 0.04
        ha = "left"
    else:
        x_text = -0.04
        ha = "right"


    plt.text(
        x_text,
        i,
        label,
        va="center",
        ha=ha,
        fontsize=8,
    )



plt.tight_layout()
plt.show()
```

```python
# Swansea players (robust match on club name)
swansea_players = players[players["club"].astype(str).str.lower().str.contains("swansea")].copy()
print("Players for Swansea:", len(swansea_players))

# Ensure per-90 contributions
if "p90_xg_xag" not in swansea_players.columns:
    if {"p90_xg", "p90_xag"}.issubset(swansea_players.columns):
        swansea_players["p90_xg_xag"] = swansea_players["p90_xg"] + swansea_players["p90_xag"]
    else:
        raise ValueError("Missing p90_xg and/or p90_xag in player stats.")

if "p90_ga" not in swansea_players.columns:
    if "p90_g+a" in swansea_players.columns:
        swansea_players["p90_ga"] = swansea_players["p90_g+a"]
    else:
        raise ValueError("Missing p90_g+a column for G+A per 90.")

if "90s" not in swansea_players.columns:
    raise ValueError("Expected '90s' column in player stats.")

# Only players with at least 3 full matches (3 x 90 mins)
swansea_players = swansea_players[swansea_players["90s"] >= 3].copy()
print("After 3+ 90s filter:", len(swansea_players))

swansea_players[
    ["player_name", "position", "90s", "g+a", "p90_xg_xag", "p90_ga"]
].sort_values("g+a", ascending=False).head(10)
```

```python
import re
import matplotlib.cm as cm

# Order used to normalise multi-role labels
ROLE_ORDER = ["GK", "DF", "MF", "FW", "Other"]

def _map_token_to_role(token: str) -> str:
    """Map a single raw token (e.g. 'AM', 'WB') to a coarse role."""
    t = token.upper()

    # Forwards
    if t.startswith(("FW", "ST", "CF")):
        return "FW"

    # Midfielders
    if t.startswith(("MF", "AM", "DM", "CM", "LM", "RM")):
        return "MF"

    # Defenders
    if t.startswith(("DF", "FB", "WB", "CB", "RB", "LB")):
        return "DF"

    # Goalkeepers
    if t.startswith("GK"):
        return "GK"

    return "Other"


def simplify_pos(pos: str) -> str:
    """
    Simplify position strings into:
      - single roles: 'FW', 'MF', 'DF', 'GK', 'Other'
      - or hybrid roles: 'FW/MF', 'DF/MF', etc.

    'MF, FW' and 'FW, MF' both become 'FW/MF' (order normalised).
    """
    if not isinstance(pos, str):
        return "Other"

    # Split on commas, slashes, semicolons, whitespace
    tokens = re.split(r"[,/; ]+", pos)
    tokens = [t for t in tokens if t]

    if not tokens:
        return "Other"

    roles = {_map_token_to_role(tok) for tok in tokens}

    # If we have recognised roles + "Other", drop "Other"
    if len(roles) > 1 and "Other" in roles:
        roles.discard("Other")

    if not roles:
        return "Other"

    # Single-role case
    if len(roles) == 1:
        return next(iter(roles))

    # Multi-role case → stable label like 'DF/MF' or 'DF/MF/FW'
    ordered = [r for r in ROLE_ORDER if r in roles]
    return "/".join(ordered)


# Ensure pos_group exists on Swansea players
swansea_players["pos_group"] = swansea_players["position"].apply(simplify_pos)

# Build a distinct colour for each pos_group (single or hybrid)
pos_groups = sorted(swansea_players["pos_group"].dropna().unique())
n_groups = len(pos_groups)

base_colors = {
    "FW": np.array([1.0, 0.2, 0.2]),
    "MF": np.array([0.2, 0.4, 1.0]),
    "DF": np.array([0.2, 1.0, 0.4]),
    "GK": np.array([0.0, 0.0, 0.0]),
    "Other": np.array([0.7, 0.7, 0.7]),
}

def blend_roles(role_label):
    roles = role_label.split("/")
    cols = [base_colors.get(r, base_colors["Other"]) for r in roles]
    return tuple(np.mean(cols, axis=0))

pos_palette = {group: blend_roles(group) for group in pos_groups}

#pos_groups, pos_palette
```

```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# Attacking map (old Cell 9)
# =========================

df_plot = swansea_players.copy()

# Exclude goalkeepers
df_plot = df_plot[df_plot["pos_group"] != "GK"]

# Top 12 outfield players by minutes played
top = df_plot.sort_values("90s", ascending=False).head(12)

# Size = minutes played (scaled to look good)
sizes_att = df_plot["90s"] * 20  # adjust multiplier as needed

colors_att = df_plot["pos_group"].map(pos_palette)

x_att = df_plot["p90_xg_xag"]
y_att = df_plot["p90_ga"]

# =========================
# Progression map (old Cell 10)
# =========================

required = ["prgc", "prgp", "90s"]
missing = [c for c in required if c not in swansea_players.columns]
if missing:
    raise ValueError(f"Missing progression columns: {missing}")

df_prog = swansea_players.copy()
df_prog = df_prog[df_prog["pos_group"] != "GK"]  # no goalkeepers

df_prog["prgc_p90"] = df_prog["prgc"] / df_prog["90s"]
df_prog["prgp_p90"] = df_prog["prgp"] / df_prog["90s"]

# Top 12 by minutes
top_prog = df_prog.sort_values("90s", ascending=False).head(12)

# Point size = minutes played
sizes_prog = df_prog["90s"] * 20

colors_prog = df_prog["pos_group"].map(pos_palette)

x_prog = df_prog["prgp_p90"]
y_prog = df_prog["prgc_p90"]

# =========================
# Side-by-side figure
# =========================

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

# ----- LEFT: Attacking contribution map (old Cell 9) -----
sc1 = ax1.scatter(
    x_att,
    y_att,
    s=sizes_att,
    c=colors_att,
    alpha=0.75,
    edgecolor="k",
    linewidth=0.5,
)

ax1.set_xlabel("Expected attacking contribution per 90 (p90_xG + p90_xAG)")
ax1.set_ylabel("Actual G+A per 90 (p90_g+a)")
ax1.set_title(f"{SWANSEA_NAME} – Player Attacking Contribution Map")
ax1.grid(True)

# 1:1 diagonal
val_min_att = min(x_att.min(), y_att.min())
val_max_att = max(x_att.max(), y_att.max())
diag_att = np.linspace(val_min_att, val_max_att, 200)
ax1.plot(diag_att, diag_att, linestyle="--", color="gray", linewidth=1)

# Label top 12 most-used players
for _, row in top.iterrows():
    ax1.text(
        row["p90_xg_xag"],
        row["p90_ga"],
        row["player_name"],
        fontsize=8,
        weight="medium",
        ha="left",
        va="bottom",
    )

# ----- RIGHT: Ball progression map (old Cell 10) -----
sc2 = ax2.scatter(
    x_prog,
    y_prog,
    s=sizes_prog,
    c=colors_prog,
    alpha=0.75,
    edgecolor="k",
    linewidth=0.5,
)

ax2.set_xlabel("Progressive passes per 90 (prgp_p90)")
ax2.set_ylabel("Progressive carries per 90 (prgc_p90)")
ax2.set_title(f"{SWANSEA_NAME} – Ball Progression Map")
ax2.grid(True)

# Diagonal (equal contribution via passes/carries)
val_min_prog = min(x_prog.min(), y_prog.min())
val_max_prog = max(x_prog.max(), y_prog.max())
diag_prog = np.linspace(val_min_prog, val_max_prog, 200)
ax2.plot(diag_prog, diag_prog, linestyle="--", color="gray", linewidth=1)

# Label top 12 most-used outfield players
for _, row in top_prog.iterrows():
    ax2.text(
        row["prgp_p90"],
        row["prgc_p90"],
        row["player_name"],
        fontsize=8,
        weight="medium",
        ha="left",
        va="bottom",
    )

# ----- Shared legend (position groups) -----
handles = []
labels = []

for group in pos_groups:
    handles.append(
        plt.Line2D(
            [0], [0],
            marker="o",
            linestyle="",
            markerfacecolor=pos_palette[group],
            markeredgecolor="k",
            markersize=6,
        )
    )
    labels.append(group)

# Put legend on the right plot (you can move it if you prefer)
ax2.legend(handles, labels, title="Position group", fontsize=8)

plt.tight_layout()
plt.show()
```

```python
df_min = swansea_players.copy()

if "min" not in df_min.columns:
    raise ValueError("Expected 'min' column with total minutes in player stats.")

df_min = df_min.sort_values(["pos_group", "min"], ascending=[True, False])

plt.figure(figsize=(9, 6))

y_labels = []
y_pos = []
bar_pos = 0

# Order groups: base roles first, then hybrids/others sorted
base_order = ROLE_ORDER  # ["GK", "DF", "MF", "FW", "Other"]
all_groups = list(df_min["pos_group"].dropna().unique())

ordered_groups = [g for g in base_order if g in all_groups] + sorted(
    [g for g in all_groups if g not in base_order]
)

for pos_group in ordered_groups:
    group = df_min[df_min["pos_group"] == pos_group]
    if group.empty:
        continue
    for _, row in group.iterrows():
        plt.barh(
            bar_pos,
            row["min"],
            color=pos_palette.get(pos_group, (0.6, 0.6, 0.6, 1.0)),
            edgecolor="k",
            alpha=0.85,
        )
        y_labels.append(f"{row['player_name']} ({pos_group})")
        y_pos.append(bar_pos)
        bar_pos += 1
    bar_pos += 0.5  # gap between groups

plt.yticks(y_pos, y_labels, fontsize=8)
plt.xlabel("Minutes played")
plt.title(f"{SWANSEA_NAME} – Minutes & Role Distribution")

plt.tight_layout()
plt.show()
```

```python
# ===========================================
# Swansea City – Automated Insights (Standard Stats Only)
# ===========================================

import numpy as np
import pandas as pd

df = swansea_players.copy()

print("### Swansea City – Data-Driven Summary (Standard Stats)\n")

# -------------------------------------------------------------------
# Ensure per-90 columns exist (based only on standard stats)
# -------------------------------------------------------------------
if "p90_xg_xag" not in df.columns:
    if {"p90_xg", "p90_xag"}.issubset(df.columns):
        df["p90_xg_xag"] = df["p90_xg"] + df["p90_xag"]
    else:
        raise ValueError("Missing p90_xg and/or p90_xag columns in standard stats.")

if "p90_ga" not in df.columns:
    if "p90_g+a" in df.columns:
        df["p90_ga"] = df["p90_g+a"]
    else:
        raise ValueError("Missing p90_g+a column for G+A per 90.")

# Make sure core columns exist
for col in ["g+a", "min", "90s"]:
    if col not in df.columns:
        raise ValueError(f"Missing required standard column: {col}")


def player_list(rows, col="player_name", n=None):
    if n is not None:
        rows = rows.head(n)
    return ", ".join(rows[col].tolist())


# ======================================================
# 1. Attacking Output vs Expected Output (p90 G+A vs p90 xG+xAG)
# ======================================================
df["delta_ga"] = df["p90_ga"] - df["p90_xg_xag"]

overperformers = df.sort_values("delta_ga", ascending=False).head(5)
underperformers = df.sort_values("delta_ga", ascending=True).head(5)

print("#### 1) Attacking Output vs Expected Output (per 90)")
print("Based on p90 G+A vs p90 xG+xAG:")

print("- Overperformers (scoring/assisting more than expected):")
print("  ", player_list(overperformers))

print("- Underperformers (return below expected xG+xAG):")
print("  ", player_list(underperformers))
print()


# ======================================================
# 2. High-Minutes Profiles vs Productivity
# ======================================================
high_minutes = df.sort_values("min", ascending=False).head(7)
high_p90_ga = df.sort_values("p90_ga", ascending=False).head(7)

# High minutes but low p90 G+A → high reliance, modest return
median_p90_ga = df["p90_ga"].median()
workhorse_low_return = (
    df[df["min"] >= df["min"].median()]
    .sort_values("p90_ga", ascending=True)
    .head(5)
)

print("#### 2) High-Minutes Players vs Attacking Productivity")
print("- Most heavily-used players by total minutes:")
print("  ", player_list(high_minutes))

print("- Best G+A per 90 (p90_g+a) among Swansea players:")
print("  ", player_list(high_p90_ga))

print("- High-minute players with relatively low p90 G+A (potential efficiency concern):")
print("  ", player_list(workhorse_low_return))
print()


# ======================================================
# 3. Progression vs Expected Contribution
#    (using standard progressive passes & carries)
# ======================================================
missing_prog = [c for c in ["prgc", "prgp"] if c not in df.columns]
if not missing_prog:
    df["progression_total"] = df["prgc"].fillna(0) + df["prgp"].fillna(0)

    top_progressors = df.sort_values("progression_total", ascending=False).head(5)

    # Strong progressors but below-median expected attacking output
    prog_thresh = df["progression_total"].median()
    xg_thresh = df["p90_xg_xag"].median()
    support_progressors = df[
        (df["progression_total"] >= prog_thresh) &
        (df["p90_xg_xag"] < xg_thresh)
    ].sort_values("progression_total", ascending=False).head(5)

    print("#### 3) Progression vs Expected Contribution")
    print("- Top progressors (progressive passes + carries):")
    print("  ", player_list(top_progressors))

    print("- Progressors whose xG+xAG per 90 is still below team median:")
    print("  ", player_list(support_progressors))
    print()
else:
    print("#### 3) Progression vs Expected Contribution")
    print(f"Progression columns missing from standard stats: {missing_prog}")
    print()


# ======================================================
# 4. Minutes & 'Impact Sub' Profiles
# ======================================================
# Players with below-median minutes but above-median p90 G+A
minutes_median = df["min"].median()
ga_median = df["p90_ga"].median()

impact_subs = df[
    (df["min"] < minutes_median) &
    (df["p90_ga"] > ga_median)
].sort_values("p90_ga", ascending=False).head(5)

print("#### 4) Minutes & 'Impact' Profiles")
print("- Most minutes played:")
print("  ", player_list(high_minutes))

print("- Low-minutes players with above-median p90 G+A (potential impact subs):")
print("  ", player_list(impact_subs))
print()


# ======================================================
# 5. Position Group Breakdown (single + hybrid)
# ======================================================
if "pos_group" in df.columns:
    pos_counts = df["pos_group"].value_counts()
    print("#### 5) Position Groups (Standard + Hybrid Roles)")
    print("Breakdown of Swansea positional profiles:")
    for pos, count in pos_counts.items():
        print(f"  - {pos}: {count} players")
    print()
else:
    print("#### 5) Position Groups")
    print("pos_group column not found; run the position grouping cell first.")
    print()

print("### Overall Summary")
print(
    "These insights summarise Swansea's attacking efficiency, reliance on key players, "
    "progression structure (from standard progressive metrics), and potential impact options "
    "off the bench – all using only standard FBRef stats."
)
```
